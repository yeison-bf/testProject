```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Variable global para almacenar el DataFrame
df_global = None

def paso_1_cargar_datos():
    """Carga el archivo CSV"""
    global df_global
    print("\n" + "="*60)
    print("üìÇ PASO 1: CARGANDO DATOS")
    print("="*60)

    archivo = input("\nIngresa el nombre del archivo CSV: ").strip()

    try:
        df_global = pd.read_csv(archivo)
        print(f"\n‚úÖ Datos cargados exitosamente!")
        print(f"   - Filas: {df_global.shape[0]}")
        print(f"   - Columnas: {df_global.shape[1]}")
        print(f"\nPrimeras columnas disponibles:")
        for i, col in enumerate(df_global.columns[:5], 1):
            print(f"   {i}. {col}")
        if len(df_global.columns) > 5:
            print(f"   ... y {len(df_global.columns) - 5} columnas m√°s")

        print(f"\n‚úÖ DataFrame almacenado en memoria correctamente")
        return True
    except FileNotFoundError:
        print(f"\n‚ùå Error: El archivo '{archivo}' no se encontr√≥")
        print("   Aseg√∫rate de que el archivo est√© en el mismo directorio")
        df_global = None
        return False
    except Exception as e:
        print(f"\n‚ùå Error al cargar: {e}")
        df_global = None
        return False

def paso_2_limpiar_datos():
    """Limpieza de datos"""
    global df_global

    if df_global is None:
        print("\n‚ùå Primero debes cargar los datos (Opci√≥n 1)")
        return

    print("\n" + "="*60)
    print("üßπ PASO 2: LIMPIANDO DATOS")
    print("="*60)

    # Convertir fecha a datetime
    df_global['Week Ending'] = pd.to_datetime(df_global['Week Ending'])
    print("\n‚úì Fechas convertidas a formato datetime")

    # Eliminar valores nulos
    antes = len(df_global)
    df_global = df_global.dropna()
    despues = len(df_global)
    print(f"‚úì Valores nulos eliminados: {antes - despues} filas")

    # Eliminar duplicados
    antes = len(df_global)
    df_global = df_global.drop_duplicates()
    despues = len(df_global)
    print(f"‚úì Duplicados eliminados: {antes - despues} filas")

    print(f"\n‚úÖ Limpieza completada!")
    print(f"   - Filas restantes: {len(df_global)}")

def paso_3_analisis_estadistico():
    """An√°lisis estad√≠stico descriptivo"""
    global df_global

    if df_global is None:
        print("\n‚ùå Primero debes cargar los datos (Opci√≥n 1)")
        return

    print("\n" + "="*60)
    print("üìä PASO 3: AN√ÅLISIS ESTAD√çSTICO")
    print("="*60)

    columnas_numericas = ['Adult-Use Retail Sales', 'Medical Marijuana Retail Sales',
                          'Total Adult-Use and Medical Sales']

    print("\nüìà ESTAD√çSTICAS DESCRIPTIVAS:\n")
    estadisticas = df_global[columnas_numericas].describe()
    print(estadisticas)

    # Calcular totales
    total_adult = df_global['Adult-Use Retail Sales'].sum()
    total_medical = df_global['Medical Marijuana Retail Sales'].sum()
    total_general = df_global['Total Adult-Use and Medical Sales'].sum()

    print(f"\nüí∞ TOTALES ACUMULADOS:")
    print(f"   - Ventas Adult-Use: ${total_adult:,.2f}")
    print(f"   - Ventas Medical: ${total_medical:,.2f}")
    print(f"   - Ventas Totales: ${total_general:,.2f}")

    # Promedios
    print(f"\nüìä PROMEDIOS:")
    print(f"   - Promedio Adult-Use: ${df_global['Adult-Use Retail Sales'].mean():,.2f}")
    print(f"   - Promedio Medical: ${df_global['Medical Marijuana Retail Sales'].mean():,.2f}")
    print(f"   - Precio prom. Adult-Use: ${df_global['Adult-Use Average Product Price'].mean():.2f}")
    print(f"   - Precio prom. Medical: ${df_global['Medical Average Product Price'].mean():.2f}")

def grafica_1_tendencia_ventas():
    """Gr√°fica de tendencia de ventas totales"""
    plt.figure(figsize=(12, 6))
    plt.plot(df_global['Week Ending'], df_global['Total Adult-Use and Medical Sales'],
             color='green', linewidth=2, marker='o', markersize=3)
    plt.title('Tendencia de Ventas Totales', fontsize=16, fontweight='bold')
    plt.xlabel('Fecha', fontsize=12)
    plt.ylabel('Ventas ($)', fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig('tendencia_ventas.png', dpi=300, bbox_inches='tight')
    print("\n‚úÖ Gr√°fica guardada como 'tendencia_ventas.png'")
    plt.show()

def grafica_2_comparacion():
    """Gr√°fica comparativa Adult-Use vs Medical"""
    plt.figure(figsize=(12, 6))
    plt.plot(df_global['Week Ending'], df_global['Adult-Use Retail Sales'],
             label='Adult-Use', color='blue', linewidth=2)
    plt.plot(df_global['Week Ending'], df_global['Medical Marijuana Retail Sales'],
             label='Medical', color='red', linewidth=2)
    plt.title('Comparaci√≥n: Adult-Use vs Medical', fontsize=16, fontweight='bold')
    plt.xlabel('Fecha', fontsize=12)
    plt.ylabel('Ventas ($)', fontsize=12)
    plt.legend(fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig('comparacion_ventas.png', dpi=300, bbox_inches='tight')
    print("\n‚úÖ Gr√°fica guardada como 'comparacion_ventas.png'")
    plt.show()

def grafica_3_productos():
    """Gr√°fica de productos vendidos"""
    plt.figure(figsize=(12, 6))
    plt.bar(df_global['Week Ending'], df_global['Total Products Sold'],
            color='purple', alpha=0.7)
    plt.title('Total de Productos Vendidos por Semana', fontsize=16, fontweight='bold')
    plt.xlabel('Fecha', fontsize=12)
    plt.ylabel('Cantidad de Productos', fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3, axis='y')
    plt.tight_layout()
    plt.savefig('productos_vendidos.png', dpi=300, bbox_inches='tight')
    print("\n‚úÖ Gr√°fica guardada como 'productos_vendidos.png'")
    plt.show()

def grafica_4_precios():
    """Gr√°fica de evoluci√≥n de precios"""
    plt.figure(figsize=(12, 6))
    plt.plot(df_global['Week Ending'], df_global['Adult-Use Average Product Price'],
             label='Adult-Use', color='orange', linewidth=2, marker='o')
    plt.plot(df_global['Week Ending'], df_global['Medical Average Product Price'],
             label='Medical', color='cyan', linewidth=2, marker='s')
    plt.title('Evoluci√≥n de Precios Promedio', fontsize=16, fontweight='bold')
    plt.xlabel('Fecha', fontsize=12)
    plt.ylabel('Precio ($)', fontsize=12)
    plt.legend(fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig('precios_promedio.png', dpi=300, bbox_inches='tight')
    print("\n‚úÖ Gr√°fica guardada como 'precios_promedio.png'")
    plt.show()

def grafica_5_todas():
    """Genera todas las gr√°ficas en un solo dashboard"""
    fig, axes = plt.subplots(2, 2, figsize=(16, 10))
    fig.suptitle('Dashboard Completo - An√°lisis Cannabis Retail', fontsize=18, fontweight='bold')

    # Gr√°fica 1: Tendencia
    axes[0, 0].plot(df_global['Week Ending'], df_global['Total Adult-Use and Medical Sales'],
                    color='green', linewidth=2)
    axes[0, 0].set_title('Tendencia Ventas Totales', fontweight='bold')
    axes[0, 0].set_xlabel('Fecha')
    axes[0, 0].set_ylabel('Ventas ($)')
    axes[0, 0].tick_params(axis='x', rotation=45)
    axes[0, 0].grid(True, alpha=0.3)

    # Gr√°fica 2: Comparaci√≥n
    axes[0, 1].plot(df_global['Week Ending'], df_global['Adult-Use Retail Sales'],
                    label='Adult-Use', color='blue', linewidth=2)
    axes[0, 1].plot(df_global['Week Ending'], df_global['Medical Marijuana Retail Sales'],
                    label='Medical', color='red', linewidth=2)
    axes[0, 1].set_title('Adult-Use vs Medical', fontweight='bold')
    axes[0, 1].set_xlabel('Fecha')
    axes[0, 1].set_ylabel('Ventas ($)')
    axes[0, 1].legend()
    axes[0, 1].tick_params(axis='x', rotation=45)
    axes[0, 1].grid(True, alpha=0.3)

    # Gr√°fica 3: Productos
    axes[1, 0].bar(df_global['Week Ending'], df_global['Total Products Sold'],
                   color='purple', alpha=0.7)
    axes[1, 0].set_title('Productos Vendidos', fontweight='bold')
    axes[1, 0].set_xlabel('Fecha')
    axes[1, 0].set_ylabel('Cantidad')
    axes[1, 0].tick_params(axis='x', rotation=45)
    axes[1, 0].grid(True, alpha=0.3, axis='y')

    # Gr√°fica 4: Precios
    axes[1, 1].plot(df_global['Week Ending'], df_global['Adult-Use Average Product Price'],
                    label='Adult-Use', color='orange', linewidth=2, marker='o')
    axes[1, 1].plot(df_global['Week Ending'], df_global['Medical Average Product Price'],
                    label='Medical', color='cyan', linewidth=2, marker='s')
    axes[1, 1].set_title('Precios Promedio', fontweight='bold')
    axes[1, 1].set_xlabel('Fecha')
    axes[1, 1].set_ylabel('Precio ($)')
    axes[1, 1].legend()
    axes[1, 1].tick_params(axis='x', rotation=45)
    axes[1, 1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('dashboard_completo.png', dpi=300, bbox_inches='tight')
    print("\n‚úÖ Dashboard guardado como 'dashboard_completo.png'")
    plt.show()

def menu_graficas():
    """Submen√∫ para seleccionar gr√°ficas"""
    global df_global

    if df_global is None:
        print("\n‚ùå Primero debes cargar los datos (Opci√≥n 1)")
        return

    while True:
        print("\n" + "="*60)
        print("üìà MEN√ö DE GR√ÅFICAS")
        print("="*60)
        print("1. Tendencia de Ventas Totales")
        print("2. Comparaci√≥n Adult-Use vs Medical")
        print("3. Productos Vendidos por Semana")
        print("4. Evoluci√≥n de Precios Promedio")
        print("5. Dashboard Completo (Todas las gr√°ficas)")
        print("0. Volver al men√∫ principal")
        print("="*60)

        opcion = input("\nSelecciona una gr√°fica: ").strip()

        match opcion:
            case '1':
                grafica_1_tendencia_ventas()
            case '2':
                grafica_2_comparacion()
            case '3':
                grafica_3_productos()
            case '4':
                grafica_4_precios()
            case '5':
                grafica_5_todas()
            case '0':
                break
            case _:
                print("\n‚ùå Opci√≥n no v√°lida. Intenta de nuevo.")

        input("\nPresiona ENTER para continuar...")

def menu_principal():
    """Men√∫ principal del sistema"""
    print("\n" + "="*60)
    print("  üåø BIENVENIDO AL AN√ÅLISIS CANNABIS RETAIL üåø")
    print("="*60)
    print("\nAseg√∫rate de tener tu archivo CSV en el mismo directorio")

    while True:
        print("\n" + "="*60)
        print("  MEN√ö PRINCIPAL")
        print("="*60)

        # Mostrar estado de los datos
        if df_global is not None:
            print(f"üìä Estado: Datos cargados ({len(df_global)} filas)")
        else:
            print("‚ö†Ô∏è  Estado: Sin datos cargados")

        print("\n1. Cargar Documento")
        print("2. Limpieza de Datos")
        print("3. An√°lisis Estad√≠stico")
        print("4. Generar Gr√°ficas")
        print("0. Salir")
        print("="*60)

        opcion = input("\nSelecciona una opci√≥n: ").strip()

        match opcion:
            case '1':
                paso_1_cargar_datos()
            case '2':
                paso_2_limpiar_datos()
            case '3':
                paso_3_analisis_estadistico()
            case '4':
                menu_graficas()
            case '0':
                print("\n" + "="*60)
                print("  üëã ¬°Hasta luego! An√°lisis finalizado.")
                print("="*60 + "\n")
                break
            case _:
                print("\n‚ùå Opci√≥n no v√°lida. Por favor elige 1, 2, 3, 4 o 0.")

        if opcion != '0':
            input("\n[Presiona ENTER para continuar...]")

# ==================== EJECUCI√ìN ====================
if __name__ == "__main__":
    menu_principal()
```
